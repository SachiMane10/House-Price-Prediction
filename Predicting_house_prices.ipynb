```python
from sklearn.model_selection import train_test_split

x= data.drop('median_house_value', axis=1)
y= data['median_house_value']
```

```python
from sklearn.linear_model import LinearRegression


x_train, y_train = train_data.drop(['median_house_value'], axis=1), train_data['median_house_value']

reg = LinearRegression()

reg.fit(x_train, y_train)
```

```python
test_data = x_test.join(y_test)

test_data['total_rooms'] = np.log(test_data['total_rooms'] + 1)
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms'] + 1)
test_data['population'] = np.log(test_data['population'] + 1)
test_data['households'] = np.log(test_data['households'] + 1)

test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(['ocean_proximity'], axis=1)

test_data['rooms_per_household'] = test_data['total_rooms'] / test_data['households']
test_data['bedrooms_per_room'] = test_data['total_bedrooms'] / test_data['total_rooms']
```

```python
x_test, y_test = test_data.drop(['median_house_value'], axis=1), test_data['median_house_value']
```

```python
x_test = x_test.drop(columns=['rooms_per_household', 'bedrooms_per_room'], errors='ignore')

x_test['bedroom_ratio'] = x_test['total_bedrooms'] / x_test['total_rooms']
x_test['household_rooms'] = x_test['total_rooms'] / x_test['households']

x_test = x_test[x_train.columns]

reg.score(x_test, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()

forest.fit(x_train, y_train)
```

```python
forest.score(x_test, y_test)
```

parameter grid

grid search with cross validation (We take the data and we split it into K folds and use all but one fold to train the data, the other for evaluation)

```python
from sklearn.model_selection import GridSearchCV

forest = RandomForestRegressor()

param_grid = {
    "n_estimators": [3, 10, 30],
    "max_features": [2, 4, 6, 8]
}

grid_search = GridSearchCV(forest, param_grid, cv=5, scoring='neg_mean_squared_error', return_train_score=True)

grid_search.fit(x_train, y_train)
```

```python
best_forest = grid_search.best_estimator_
```

```python
best_forest.score(x_test, y_test)
```
